```python
sns.countplot(data=df, x='Churn')
plt.title("Customer Churn Count")
plt.savefig('../output/crunch_count.png')
plt.show()
```

```python
#Defining function to add borders around plots

def add_border(ax, color='black', linewidth=1.5):
   
   for spine in ax.spines.values():
        spine.set_visible(True)
        spine.set_edgecolor(color)
        spine.set_linewidth(linewidth)
```

```python
#Churn by Binary Demographics

binary_cols = ['gender', 'SeniorCitizen', 'Partner', 'Dependents']

fig, axes = plt.subplots(2, 2, figsize=(18, 10))

for i, cols in enumerate(binary_cols):
    row = i // 2
    col = i % 2
    sns.countplot(data=df, x=cols, hue='Churn', ax=axes[row][col])
    axes[row][col].set_title(f"Churn by {cols}")
    
    add_border(axes[row][col])

plt.tight_layout()
fig.savefig('../output/crunch_by_demographics.png')
plt.show()
```

```python
#Churn by Service Type

service_cols = ['PhoneService', 'MultipleLines', 'InternetService',
                'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
                'TechSupport', 'StreamingTV', 'StreamingMovies']


fig, axes = plt.subplots(3, 3, figsize=(18, 10))

for i, cols in enumerate(service_cols):
    row = i // 3
    col = i % 3
    sns.countplot(data=df, x=cols, hue='Churn', ax=axes[row][col])
    axes[row][col].set_title(f"Churn by {cols}")
    
    add_border(axes[row][col])

plt.tight_layout()
fig.savefig('../output/churn_by_service_type.png')
plt.show()
```

```python
#Churn by Contract & Payment Method

sns.countplot(data=df, x='Contract', hue='Churn')
plt.title("Churn by Contract Type")
plt.savefig('../output/churn_by_contract.png')
plt.show()

sns.countplot(data=df, x='PaymentMethod', hue='Churn')
plt.title("Churn by Payment Method")
plt.xticks(rotation=45)
plt.savefig('../output/churn_by_payment_method.png')
plt.show()
```

```python
#Churn vs. Tenure and Monthly Charges

plt.figure(figsize=(14, 6))

# Tenure distribution
plt.subplot(1, 2, 1)
sns.histplot(data=df, x='tenure', hue='Churn', bins=30, kde=True)
plt.title("Churn vs. Tenure")
add_border(plt.gca())


# MonthlyCharges boxplot
plt.subplot(1, 2, 2)
sns.boxplot(x='Churn', y='MonthlyCharges', data=df)
plt.title("Monthly Charges vs. Churn")
add_border(plt.gca())

plt.tight_layout()
fig.savefig('../output/churn_by_tenure_and_monthly_charges.png')
plt.show()
```

```python
df['Churn_Encoded'] = df['Churn'].map({'No': 0, 'Yes': 1})
```

```python
numeric_df = df[['tenure', 'MonthlyCharges', 'TotalCharges', 'SeniorCitizen', 'Churn_Encoded']]

corr_matrix = numeric_df.corr()

plt.figure(figsize=(6, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap of Numerical Features")
plt.savefig('../output/correlation.png')
plt.show()
```

### 🔍 Key Insights from Exploratory Data Analysis

**Churn Overview:**
- 26.5% of customers have churned — a significant rate to explore further.

**Demographics:**
- Senior citizens are more likely to churn.
- Customers without partners or dependents also show higher churn rates.
- Gender has no clear impact on churn.

**Services:**
- Lack of services like Tech Support, Online Security, and Backup correlates with higher churn.
- Customers with Fiber Optic internet are more likely to churn than DSL users.
- Having multiple lines or streaming services shows mixed churn behavior.

**Contract & Billing:**
- Month-to-month contracts are strongly associated with higher churn.
- Customers on longer-term contracts (1 or 2 years) are more loyal.
- Electronic check users churn the most.

**Numerical Features:**
- `tenure` has the strongest (negative) correlation with churn (-0.35).
- Higher `MonthlyCharges` is weakly associated with churn.
- `TotalCharges` correlates negatively with churn — loyal customers pay more over time.
